# Inheritance

Inheritance is the capability of one class to derive or inherit the properties from some another class. The benefits of inheritance are:
1.	It represents real-world relationships well.

2.	It provides reusability of a code. We donâ€™t have to write the same code again and again. Also, it allows us to add more features to a class without modifying it.

3.	It is transitive in nature, which means that if class B inherits from another class A, then all the subclasses of B would automatically inherit from class A.


# Types of Inheritance:

## 1. Single inheritance:

 When a child class inherits from only one parent class, it is called as single inheritance. We saw an example above.

singleih.png

```
class Person:
      variable=10
      def __init__(self,name):
          self.name=name
      def display(self):
          print("you are call display")
          print(self.address)
class Employee(Person):
    def __init__(self,address):
        self.address=address
p=Person("jill")
e=Employee("H.no.119, nehru nagar, Noida,UP")
e.display()
print(e.variable)
```

## 2. Multiple inheritance: 

When a child class inherits from multiple parent classes, it is called as multiple inheritance.

multiih.png

```
class A:
    def __init__(self,name1):
        self.name1=name1
class B:
    def __init__(self,name2):
        self.name2=name2
class C(A,B):
    def __init__(self,name1,name2):
        A.__init__(self,name1)
        B.__init__(self,name2)
    def display(self):
        print(self.name1,self.name2)
obj=C("jack","rahul")
obj.display()
```

## 3.	Multilevel inheritance: 

When we have child and grand child relationship

multilevelih.png

```
class A:
    def __init__(self,name):
        self.name=name
    def getName(self):
        print(self.name)
class B(A):
    def __init__(self,name,age):
        A.__init__(self,name)
        self.age=age
    def getAge(self):
        print(self.age)
class C(B):
    def __init__(self,name,age,address):
        B.__init__(self,name,age)
        self.address=address
    def getAddress(self):
        print(self.address)
obj=C("jack",23,"H.no.119, nehru nagar, Noida,UP")
obj.getAddress()
obj.getAge()
obj.getName()
```

## 4. Hierarchical inheritance 

More than one derived classes are created from a single base.

Hierarchicalih.png

```
class A:
    def funA(self):
        print("this fun is in class A")
class B(A):
    def funB(self):
        print("this fun is in class B")
class C(A):
    def funC(self):
        print("this fun is in class C")

class D(A):
    def funD(self):
        print("this fun is in class D")
a=A()
b=B()
c=C()
d=D()
b.funA()
c.funA()
d.funA()
```

## 5.Hybrid Inheritance:

Inheritence consisting of multiple types of inheritence is called hybrid inheritence.

hybridih.png

```
class A:
    def fun(self):
        print("this fun is in class A")
class B(A):
    def fun1(self):
        print("this fun is in class B")
class C(A):
    def fun1(self):
        print("this fun is in class C")

class D(B,C):
    def fun3(self):
        print("this fun is in class D")
obj=D()
obj.fun1()
```

# super function

The super() function is used to give access to methods and properties of a parent or sibling class.

The super() function returns an object that represents the parent class.

```
class C:
    name="i am in class A variable"
    def __init__(self):
        self.name="i am in class A constructor"
       
class B(C):
    name="i am in class B variable"
    def __init__(self):
        super().__init__()
        #self.name="i am in class B constructor"
b=B()
print(b.name)
```

## Advantage of using super()

```
class Mammal:
  def __init__(self, mammalName):
    print(mammalName, 'is a warm-blooded animal.')
    
class Dog(Mammal):
  def __init__(self,name):
    print('Dog has four legs.')
    super().__init__(name)
    
d1 = Dog("Dog")
```

Since we do not need to specify the name of the base class when we call its members, we can easily change the base class name (if we need to).

```
class Animal:
  def __init__(self, mammalName):
    print(mammalName, 'is a warm-blooded animal.')
    
class Dog(Animal):
  def __init__(self,name):
    print('Dog has four legs.')
    # no need to change this
    super().__init__(name)
    
d1 = Dog("Dog")
```
