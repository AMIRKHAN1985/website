# **1.Function Arguments**

# **a).Default Arguments**

```python
def greet(name, msg="good morning"):
    print(name ,msg)



greet("Kate")
greet("Bruce", "How do you do?")
```

# **b).Keyword Arguments**

```python
def greet(name, msg):
    print(name,msg)
    
# 2 keyword arguments (order)  
greet(name="rahul",msg="good morning",)
```

```python
def greet(name, msg):
    print(name,msg)
    
# 2 keyword arguments (out of order)
greet(msg = "How do you do?",name = "Bruce") 

```

```python
def greet(name, msg):
    print(name,msg)
    
#1st positional, 2nd keyword argument (This will work)
greet("Bruce", msg = "How do you do?")
```

```python
def greet(name, msg):
    print(name,msg)
    
#1st keyword argument, 2nd positional (This will give syntax error)
greet(msg = "How do you do?","Bruce")
```

# **c). Arbitrary Arguments**



```python
def greet(*names):
    # names is a tuple with arguments
    for name in names:
        print("Hello", name)


greet("Monica", "Luke", "Steve", "John")
```

# **2. Calling of Function:**

## **a).Call By Value:**

<p align="justify">In this parameter passing method, values of actual parameters are copied to functions formal parameters and the two types of parameters are stored in different memory locations. So any changes made inside functions are not reflected in actual parameters of caller.

```python
def test(string): 
    string = string+"it"
    print("id of string after passing to function",id(string))
    print("Inside Function:", string) 

string = "hi"
print("id of string before passing to function",id(string))      
# Driver's code 
test(string) 
print("Outside Function:", string) 
```

# **b). Call by Reference:**

<p align="justify">Both the actual and formal parameters refer to same locations, so any changes made inside the function are actually reflected in actual parameters of caller.

```python
def add_more(mylist): 
    mylist.append(50) 
    print("id of list after passing to function",id(mylist))
    print("Inside Function", mylist) 
  
# Driver's code 
mylist = [10,20,30,40] 
print("id of string before passing to function",id(mylist)) 
add_more(mylist) 
print("Outside Function:", mylist) 
```

# **3. Variable length arguments**

# a). *args()

<p align="justify">The special syntax *args in function definitions in python is used to pass a variable number of arguments to a function. It is used to pass a non-keyworded, variable-length argument list.

```python
def myFun(*a):
    print(type(a))
    print(a)  
    for i in a:  
        print (i) 
    
myFun('Hello', 'Welcome', 'to', 'Kanpur')  
```

```python
def myFun(a,*argv):
    print(a)
    print(type(argv))
    print(argv)  
    for arg in argv:  
        print (arg) 
    
myFun('Hello', 'Welcome', 'to', 'Kanpur')  
```

```python
def myFun(a,*argv):
    print(a)
    print(type(argv))
    print(argv)  
    for arg in argv:  
        print (arg) 
    
myFun('Hello', 'Welcome', 'to', 'Kanpur')  
```

```python
def myFun(*argv,a):  # this will not work , positional argument must come first before arbitary argument
    print(a)
    print(type(argv))
    print(argv)   
    for arg in argv:  
        print (arg) 
    
myFun('Hello', 'Welcome', 'to', 'Kanpur')  
```

```python
def myFun(*argv,a):  # this will work 
    print(a)
    print(type(argv))
    print(argv)   
    for arg in argv:  
        print (arg) 
    
myFun('Hello', 'Welcome', 'to', 'Kanpur',a="up")  
```

# b).**kwargs 

<p align="justify">The special syntax **kwargs in function definitions in python is used to pass a keyworded, variable-length argument list. We use the name kwargs with the double star. The reason is because the double star allows us to pass through keyword arguments (and any number of them).

```python
def myFun(**kwargs):
    print(type(kwargs)) 
    print(kwargs) 
    for key, value in kwargs.items(): 
        print ("%s:%s" %(key, value)) 
  
# Driver code 
myFun(amir="hello",rahul="hi")     
```

```python
def myFun(a,**kwargs):
    print(a)
    print(type(kwargs))  
    for key, value in kwargs.items(): 
        print ("%s:%s" %(key, value)) 
  
# Driver code 
myFun(5,first ='rahul', mid ='happy', last='rahim') 
```

```python
def myFun(**kwargs,a): # this will not work , positional argument must come first before arbitary argument
    print(a)
    print(type(kwargs))  
    for key, value in kwargs.items(): 
        print ("%s:%s" %(key, value)) 
  
# Driver code 
myFun(first ='rahul', mid ='happy', last='rahim',a=4) 
```

# **5. Higher order function**

<p align="justify">A function is called Higher Order Function if
it contains other functions as a parameter or
returns a function as an output i.e , the
functions that operate with another function
are known as Higher order Functions

# **Properties of higher order functions:**

A function is an instance of the Object type.

You can store the function in a variable.

You can pass the function as a parameter to another function.

You can return the function from a function.

You can store them in data structures such as hash tables, lists.

<p align="justify">In Python, a function can be assigned to a variable. This assignment does not call the function, instead a reference to that function is created.

```python
def square(x):
    return x*x
print(square)
print(square(3))
f=square
print(f)
print(f(3))
```

<p align="justify">Functions are like objects in Python, therefore, they can be passed as argument to other functions.

```python
def sum(n, square):
    total=0
    for num in range(1,n+1):
        total = total + square(num)
    return total
def square(x):
    return x*x
print(sum(3,square))
```

<p align="justify">As functions are objects, we can also return a function from another function.

```python
def make_adder (x):
    def adder(y):
        return x + y
    return adder
addition = make_adder(15)
print(addition(10))
```

# **map() Function**

<p align="justify">Map function takes another function as a parameter along with a sequence of iterables and returns an output after applying the function to each iterable present in the sequence.

SYNTAX: 

map(function,iterables)

```python
def newfunc (a):
    return  a*a
#x is the map object
x = map(newfunc , (1, 2 ,3, 4 ))
print(x)
print(set(x))
```

```python
tup= (5 , 7 , 22 , 97 , 54 , 62 , 77 , 23 , 73 , 61 )
newtuple = tuple(map(lambda x: x+ 3 , tup ))
print( newtuple)
```

# **filter() function**

<p align="justify">The filter() function is used to create an output list consisting of values for which the function returns true.

SYNTAX:

filter(function,iterables)

```python
def func(x):
    if x< 3:
        return x
y = filter(func , (1,2,3,4))
print(y)
print(list(y))
```

```python
y = filter(lambda x: (x>= 3 ), (1 ,2 ,3 ,4 ))
print(list(y))
```

# **reduce() function**

<p align="justify">The reduce() function, as the name describes, applies a given function to the iterables and returns a single value.

SYNTAX:

reduce(function,iterables)

```python
from functools import reduce
result = reduce((lambda x, y: x * y), [1, 2, 3, 4,5])
print(result)
```

# **6. Local and Global variable**

# **a).Global Variables**

<p align="justify">In Python, a variable declared outside of the function or in global scope is known as a global variable. This means that a global variable can be accessed inside or outside of the function.

```python
x = 5

def foo():
    print("x inside:", x)

foo()
print("x outside:", x)
```

```python
x = 5

def foo():
    x = x * 2  # this will give you error
    print(x)

foo()
print(x)
```

# **b).Local Variables**

<p align="justify">A variable declared inside the function's body or in the local scope is known as a local variable.

```python
def foo():
    y = 5


foo()
print(y)
```

```python
def foo():
    y = 5
    print(y)

foo()
```

**Using Global and Local variables in the same code**

```python
x = 5

def foo():
    global x
    y = 6
    x=1
    x = x * 2
    print(x)
    print(y)

foo()
print(x)
```

```python
x = 5

def foo():
    x = 10
    print("local x:", x)


foo()
print("global x:", x)
```
